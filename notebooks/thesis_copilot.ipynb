```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

from src.collect import collect_papers
from src.extract import extract_record
from src.organize import records_to_df, group_by, filter_records, top_by_confidence
from src.brief import generate_brief
```

## Step 1 — Collect Papers from PubMed


```python
query = "ovarian aging menopause delay intervention"
papers = collect_papers(query, max_results=10)
print(f"Fetched {len(papers)} papers")
papers[0]
```

## Step 2 — Extract Structured Claims


```python
# Set your Anthropic API key
os.environ['ANTHROPIC_API_KEY'] = 'your-api-key-here'

records = [extract_record(p) for p in papers]
print(f"Extracted {len(records)} records")
```

## Step 3 — Organize by Mechanism / Modality / Evidence


```python
df = records_to_df(records)
df[['title', 'target', 'pathway', 'modality', 'effect_direction', 'confidence_score']]
```

```python
# Group by modality
by_modality = group_by(records, 'modality')
for modality, recs in by_modality.items():
    print(f"{modality}: {len(recs)} papers")
```

```python
# Top records by confidence
top = top_by_confidence(records, n=5)
for r in top:
    print(f"{r.confidence_score:.2f} | {r.title[:80]}")
```

## Step 4 — Generate One-Page Decision Brief


```python
brief = generate_brief(top_by_confidence(records, n=10))
print(brief)
```
